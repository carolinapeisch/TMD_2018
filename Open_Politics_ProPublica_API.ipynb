```python
# downloaded from https://www.congress.gov/help/field-values/member-bioguide-ids
member_ids = pd.read_csv('congress_members_bioguide.csv')
```

```python
def find_member_id(name):
    x = member_ids[member_ids['LAST_NAME'].str.strip(r'\t').str.match(name)]['BIOGUIDE_ID']
    return str(x.values[0])
```
